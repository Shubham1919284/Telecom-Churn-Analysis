```python
#Importing Required Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import importlib
```

```python
#Loading the Dataset and Performing Basic operations
df=pd.read_csv("Customer Churn.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['TotalCharges']=df['TotalCharges'].replace(" ",0)
df['TotalCharges']=df['TotalCharges'].astype('float')
```

```python
df.info()
```

```python
df.isnull().sum().sum()  #This has combined all the rows and columns and checked if there is any 'NULL' 
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df['SeniorCitizen']=df['SeniorCitizen'].replace(0,'No')
df['SeniorCitizen']=df['SeniorCitizen'].replace(1,'Yes')
```

```python
df.head()
```

```python
ax = sns.countplot(x="Churn", data=df)  
ax.bar_label(ax.containers[0]) 
plt.title("Count Of Customer by Churn")  
print(plt.title)  
plt.show()
```

#This shows how many Customers are still using it and how many has left.

```python
gb=df.groupby(["Churn"]).agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct="%1.1f%%")
plt.title("Churn Distribution")
plt.show()
```

```python
ax=sns.countplot(x="gender",hue="Churn",data=df)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Churned Customer based on Gender ")  
plt.show()
```

#This shows how many have churned based on Gender.

```python
ax=sns.countplot(x="SeniorCitizen",hue="Churn",data=df)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Count Of SeniorCitizen by Churned ")
plt.show()
```

```python
ax=sns.histplot(x="tenure",data=df,hue="Churn",bins=72)
plt.title("Count Of People Churned During Time")
plt.show()
```

```python
ax=sns.countplot(x="Contract",data=df,hue="Churn")
for container in ax.containers:
    ax.bar_label(container)
plt.title("Count Of Customer by Contract")
plt.show()
```

```python
df.columns.values
```

```python
plt.figure(figsize=(10, 6)) 
columns_to_plot = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
    'PaperlessBilling', 'PaymentMethod'
]
n_cols = 4 
n_rows = -(-len(columns_to_plot) // n_cols)  
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 5)) 

axes = axes.flatten()

for i, column in enumerate(columns_to_plot):
    ax = axes[i]
    sns.countplot(x=column, data=df, hue="Churn", ax=ax)  
    ax.set_title(f"Count Plot for {column}") 
    ax.tick_params(axis='x', rotation=45)  
    for container in ax.containers:
        ax.bar_label(container)  

for j in range(len(columns_to_plot), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()  
plt.show()
```

```python
ax=sns.countplot(x="PaymentMethod",data=df,hue="Churn")
for containers in ax.containers:
    ax.bar_label(containers)
plt.xticks(rotation=45)
plt.title("Count Of Customer by PaymentMethod")
plt.show()
```
